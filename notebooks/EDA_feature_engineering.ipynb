```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

## Data Cleaning

```python
## missing values
dataset[dataset.isnull().any(axis=1)]
```

 The dataset is converted into two sets based on Region from 122th index,we can make new columns based in the Region

1 : 'Bejaija Region Dataset'

2 : 'Sidi-Bel Abbes Region Dataset' 

Add new columns with region

```python
dataset.loc[:122,'Region']=0
dataset.loc[122:,'Region']=1
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Remove the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
#Remove the 122nd row
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
##Fix spaces in column names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

## Change the required columns as integer

```python
df[['day','month','year','Temperature','RH','Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH','Ws']].astype(int)
```

```python
df.info()
```

#### Changing the other columns to float datatype

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
objects
```

```python
df.head()
```

```python
df.describe()
```

```python
## Save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```

## Exploratory Data Analysis

```python
#Drop day month and year
df_copy=df.drop(['day','month','year'],axis=1)
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
## Encoding of the categories in classes
df_copy['Classes']=df_copy['Classes'].astype(str)
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1 )
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
## Plot density plot for all features

plt.style.use('ggplot')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
## Percentage for pie chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
# plotting pie chart
classlables=["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlables,autopct='%1.1f%%')
plt.title("Pie chart of classes")
plt.show()
```
